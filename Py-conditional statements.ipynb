# Conditional statement :
- Conditional statements in Python are used to execute certain blocks of code based on specific conditions. 
- These statements help control the flow of a program, making it behave differently in different situations

# 1. If Conditional Statement :
- If statement is the simplest form of a conditional statement.
- It executes a block of code if the given condition is true.



```python
age = 20

if age >= 18:
    print('Eligible for Vote')
```

## short hand if:

```python
age = 20
if age >= 18: print('Eligible for Vote')
```

# 2. If else Conditional Statement :
- The condition in if statement is True then it will excute
- And if condition is False then else statement will excute 

```python
age = 12
if age <= 12:
    print('Free for travel')
else:
    print('pay for Ticket')
```

## short hand if else:

```python
marks = 33
result = 'Pass' if marks >= 33 else 'Fail'
print(f"You got {marks} : {result} ")
```

# 3. elif Statement
- elif statement in Python stands for "else if."
- It allows us to check multiple conditions, providing a way to execute different blocks of code based on which condition is true.
- Using elif statements makes our code more readable and efficient by eliminating the need for multiple nested if statements.

```python
age = 23

if age <= 12:
    print("child")
elif age <= 19:
    print("Teenage")
elif age <= 34:
    print("Young adult")
else:
    print("Adult")
```

# 4. Nested if..else Statement:
- Nested if..else means an if-else statement inside another if statement.
- We can use nested if statements to check conditions within conditions.

```python
age = 43
is_member = True

if age >= 60:
    if is_member:
        print('30% senior discount')
    else:
        print('20% senior discount')
else :
    print('No discounts')
```

# 5. Ternary Statement  : 

```python
age = 34
s = 'Adult' if age >=20 else 'Minor'
print(s)
```

```python
n = 5
res = 'Even' if n % 2 == 0 else 'odd'
print(res)
```

# 6. Match-Case Statement :
- match-case statement is Python's version of a switch-case found in other languages. 
- It allows us to match a variable's value against a set of patterns

```python
number = 1

match number:
 case 1:
  print("one")
 case 2 | 3:
  print("Two or three")
 case _:
  print("Other number")
```
