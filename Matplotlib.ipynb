```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x1 = np.linspace(0, 10, 50)

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
plt.show()
```

```python
plt.plot(x1, np.tan(x1), '--')
plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
plt.grid(True)
plt.show()
```

```python
# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1), '*')
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, np.sin(x1))
plt.title("Sine Function")

plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1))
plt.title("Cosine Function")

plt.tight_layout()  # overlap 
plt.show()
```

# State-Machine Interface 

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'Quadratic')
plt.plot(x, x**3, label = 'Cubic')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Simple  Plot')
plt.legend()
plt.show()
```

# Object-Oriented API

```python
fig, ax = plt.subplots(2) # it create 2 different panels within one figure

ax[0].plot(x1, np.sin(x1), 'b-') # top panel
ax[1].plot(x1, np.cos(x1), 'b-') # bottom panel, 'b-' blue color with solid line
plt.show()
```

## Objects and refferences 

```python
fig = plt.figure() # it creates a empty drawing area(plotting area)
x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

main_axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #[left, right, width, height]
inset_axes = fig.add_axes([0.5, 0.5, 0.3, 0.3])

main_axes.plot(x2, y2, 'r')
inset_axes.plot(x2, y2, 'g')

main_axes.set_xlabel('x2')
main_axes.set_ylabel('y2')
main_axes.set_title('title')
plt.show()
```

# Error bar chart 

```python
x9 = np.arange(0, 4, 0.2)
y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9, y9, yerr = e1, fmt = '.-')
plt.grid(True)
plt.show();
```

```python
x9 = np.arange(0, 4, 0.2)
y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9, y9, yerr=e1, fmt='.-', capsize=5, color='teal')
plt.title('Exponential Decay with Error Bars')
plt.xlabel('X-axis')
plt.ylabel('exp(-x)')
plt.grid(True)
```

# stacked bar chart

```python
A = [15., 30., 45., 22.]
B = [15., 25., 50., 20.]
z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)
plt.bar(z2, A, color='b', label='A')
plt.bar(z2, B, color='r', bottom=A, label='B')

plt.xticks(z2, ['Cat1', 'Cat2', 'Cat3', 'Cat4'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
#plt.grid(True)
plt.legend()
plt.show()
```

# pie chart 

```python
plt.figure(figsize=(7,7)) # creates draw area with size w7, h7

x10 = [35, 25, 20, 20]
labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
colors = ['skyblue', 'orange', 'lightgreen', 'pink']

plt.pie(
    x10,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',    # show percentage
    startangle=90,        # rotate start point
    shadow=True,          # add 3D shadow
    explode=(0.05, 0, 0, 0)  # slightly pull out the first slice
)

plt.title('Department Distribution')
plt.show()
```

# contour plot 

```python
# Generate random Z data
Z = np.random.rand(10, 20)

# Create coordinates for X (columns) and Y (rows)
x = np.arange(Z.shape[1])  # 0 ... 19
y = np.arange(Z.shape[0])  # 0 ... 9
X, Y = np.meshgrid(x, y)

# Contour plot
cp = plt.contour(X, Y, Z)
plt.xlabel('X (column)')
plt.ylabel('Y (row)')
plt.title('Contour Plot of Random Matrix')
plt.show()
```
