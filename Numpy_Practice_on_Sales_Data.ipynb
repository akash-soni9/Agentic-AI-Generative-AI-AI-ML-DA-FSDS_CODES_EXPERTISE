## Advance Operation with bussiness examples :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])
```

```python
print('sales data shape :',sales_data.shape)
print('sample data 1st 3 restaurant :\n',sales_data[:3])
```

```python
# Total sale per year 

print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:,1:], axis=0)  # axis = 0 means column wise sum
print(yearly_total)
```

```python
# Minimum sale per restaurant
min_sale = np.min(sales_data[:, 1:], axis= 1) # finds minimum row wise
print(min_sale)
```

```python
# maximum sale per year 
max_sale = np.max(sales_data[:, 1:], axis=0)
max_sale
```

```python
# average sale per restaurant 
avg_sale = np.mean(sales_data[:, 1:], axis=1, dtype= int)
print(avg_sale)
```

```python
# cummulative sale 
cum_sale = np.cumsum(sales_data[:, 1:], axis=1)
print(cum_sale)
```

```python
# plotting the sales
plt.figure(figsize=(8,6))
plt.plot(np.mean(cum_sale, axis=0))
plt.title('Average cummulative sales accross all restaurant')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
vect1 = np.array([1,2,3,4])
vect2 = np.array([4,5,6,7])
```

```python
print('vector addition :',vect1 + vect2)
print('multipliction vector :', vect1 * vect2)
print('dot product :', np.dot(vect1, vect2))
```

```python
angle = np.arccos(np.dot(vect1, vect2)/(np.linalg.norm(vect1) * np.linalg.norm(vect2)))
print(angle)
```
